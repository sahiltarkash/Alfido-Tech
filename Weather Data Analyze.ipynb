```python
sns.set_palette('dark')
sns.displot(weather_data['Temp_C'],kde =True)
```

Explaination
The above graph shows the most common temprature dgree in the year is range between 0-3 is min and 20-30 is warm weather 
The lowest temp. is -23 to -18 degrees fall under the category of freezing temp.

                      # 2nd Graph "Tempratue Variation"

```python
plt.figure(figsize=(20,10))
plt.plot(weather_data['Date/Time'],weather_data['Temp_C'],marker = 'o',color = 'orange')
plt.xlabel('Date/Time')
plt.ylabel('Temp_C')
plt.title('Temprature Variation')
plt.show()
```

```python
sns.boxplot(x='Month',y='Rel Hum_%',data=weather_data,palette='pastel')
plt.show()
```

  We can see the highest average from months
  

```python
sns.stripplot(x = 'Month',y = 'Wind Speed_km/h',
             data=weather_data,palette='dark')
plt.show()
```

From the above graph we can see that the 1st hight wind speed is in month 1(jan)

```python
sns.pointplot(x= 'Month',y= 'Visibility_km', data = weather_data)
plt.show()
```

```python
plt.figure(figsize=(8,10))
sns.boxplot(x = 'Temp_C', y='Weather',data=weather_data,palette='deep',)
plt.show()
```

```python
sns.lineplot(x ='Month', y = 'Press_kPa' , data = weather_data)
plt.xlabel('Month')
plt.ylabel("Press_kPa")
plt.title("Atmospheric pressure over time")
plt.show()
```
